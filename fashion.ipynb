```python
import matplotlib.pyplot as plt
import pandas as pd 
import numpy as np
from keras.datasets import fashion_mnist

(train_images, train_labels),(test_image, test_labels) = fashion_mnist.load_data()
```

```python
X = test_image
y = test_labels
```

```python
train_images.shape
```

```python
digit = train_images[241]  
plt.imshow(digit, cmap=plt.cm.binary) 
plt.show()
print('Vật này là: ', train_labels[241])
```

```python
for i in range(36):
  plt.subplot(6,6,i+1)
  plt.tight_layout()
  plt.imshow(train_images[i],cmap='gray',interpolation='none' )
  plt.title('So: {}'.format(train_labels[i]))
  plt.xticks([])
  plt.yticks([])
```

```python
from keras.models import Sequential
from keras.layers import Dense 
model = Sequential() 
model.add(Dense(512,activation = 'relu', input_shape = (784,)))  
```

```python
model.add(Dense(10, activation = 'Softmax'))
model.summary()
```

```python
train_images = train_images.reshape(60000,28*28)
train_images = train_images.astype('float32')/225
```

```python
test_image = test_image.reshape(10000,28*28)
test_image = test_image.astype('float32')/225
```

```python
from keras.utils import to_categorical
train_labels = to_categorical(train_labels,10)
test_labels = to_categorical(test_labels,10)
```

```python
train_labels.shape
```

```python
test_labels.shape
```

```python
model.compile(optimizer='rmsprop',loss ='categorical_crossentropy', metrics=['accuracy'])
model.fit(train_images,train_labels,epochs=10, batch_size=128)
test_loss,test_acc = model.evaluate(test_image,test_labels)
print("Mat Mat:",test_loss)
print("Do chinh xac:",test_acc)
```
